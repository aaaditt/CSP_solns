## Section 1 — Variables & Data Types

A **variable** is a named container that holds a value.
Python has several built-in **data types**:

| Type | Example | What it stores |
|------|---------|----------------|
| `int` | `18` | Whole numbers |
| `float` | `88.5` | Decimal numbers |
| `str` | `"Aadit"` | Text |
| `bool` | `True` | True or False only |

Python figures out the type automatically — you don't have to declare it like in Java/C++.

````python
# The = sign ASSIGNS a value to a variable (it does NOT mean 'equals' like in maths)
age = 18         # int — whole number
score = 88.5     # float — decimal number
name = "Aadit"   # str — text, use single or double quotes
is_student = True  # bool — only True or False (capital T and F)

# print() shows a value in the output
print("Name:", name)
print("Age:", age)
print("Score:", score)
print("Is student?", is_student)

# type() tells you what type a variable is
print("\nType of score:", type(score))  # \n adds a blank line
print("Type of name:", type(name))
````

### f-strings — The Modern Way to Mix Variables Into Text

Put an `f` before the quote, then wrap variable names in `{curly braces}`.
Python replaces `{name}` with the actual value of the variable.

Old way (ugly): `"Hello " + name + " you are " + str(age)`  
New way (clean): `f"Hello {name} you are {age}"`

````python
# f-strings — prefix the string with f, then use {variable} inside
greeting = f"Hello, my name is {name} and I am {age} years old."
print(greeting)

# You can even put expressions (calculations) inside the braces
print(f"Next year I will be {age + 1} years old.")
print(f"My score rounded to 0 decimals: {round(score)}")
````

````python
# String operations — things you can do to text
subject = "data science"

print(subject.upper())          # → DATA SCIENCE
print(subject.title())          # → Data Science (capitalize each word)
print(subject.replace("data", "computer"))  # swap a word
print(len(subject))             # count characters → 12
print(subject.split(" "))       # split into a list: ['data', 'science']
print("  spaces  ".strip())     # remove leading/trailing spaces → 'spaces'
````

````python
# Math operators
a = 10
b = 3

print(a + b)   # 13  — addition
print(a - b)   # 7   — subtraction
print(a * b)   # 30  — multiplication
print(a / b)   # 3.333... — division (always returns float)
print(a // b)  # 3   — FLOOR division (drops decimal, integer result)
print(a % b)   # 1   — MODULO (remainder after dividing: 10 / 3 = 3 remainder 1)
print(a ** b)  # 1000 — EXPONENT (10 to the power of 3)
````

## Section 2 — Lists

A **list** stores multiple values in order, inside square brackets `[ ]`.
Lists can hold any type — numbers, strings, even other lists.

**Indexing** — accessing items by position:
- Index starts at **0** (not 1)
- `fruits[0]` → first item
- `fruits[-1]` → last item (negative counts from the end)
- `fruits[1:3]` → items at index 1 and 2 (slice — the end index is NOT included)

````python
fruits = ["apple", "banana", "mango", "orange", "grapes"]
#          index 0   index 1   index 2   index 3   index 4

print(fruits[0])      # apple  (first)
print(fruits[-1])     # grapes (last)
print(fruits[1:3])    # ['banana', 'mango'] — index 1 up to (not including) 3
print(fruits[2:])     # from index 2 to the end
print(fruits[:2])     # from start up to (not including) index 2

scores = [88, 92, 75, 95, 60, 83, 70, 98, 55, 87]
print("\nLength:", len(scores))   # count of items
print("Max:", max(scores))
print("Min:", min(scores))
print("Sum:", sum(scores))
````

````python
# Modifying lists — lists are MUTABLE (can be changed after creation)
fruits = ["apple", "banana", "mango"]

fruits.append("pineapple")      # add to END
print("After append:", fruits)

fruits.insert(1, "strawberry")  # insert at index 1, shifts everything right
print("After insert:", fruits)

fruits.remove("banana")         # remove first occurrence of this value
print("After remove:", fruits)

popped = fruits.pop()           # remove AND return the last item
print("After pop:", fruits)
print("Popped item was:", popped)
````

````python
# List comprehension — build a new list using a compact one-liner
# Syntax: [expression for variable in iterable if condition]

# Long way (using a loop):
squares_long = []
for x in range(1, 6):
    squares_long.append(x ** 2)

# Short way (comprehension) — same result, one line:
squares = [x**2 for x in range(1, 6)]

print("Squares:", squares)   # [1, 4, 9, 16, 25]

# With a condition — only even numbers
evens = [x for x in range(1, 21) if x % 2 == 0]
print("Evens:", evens)

# Transform strings
names = ["aadit", "priya", "rohan"]
upper_names = [n.upper() for n in names]
print("Upper:", upper_names)
````

## Section 3 — Loops

A **loop** repeats code multiple times.

- **`for` loop** — repeat for each item in a sequence (list, range, string...)
- **`while` loop** — repeat as long as a condition is True

`range(start, stop, step)` generates a sequence of numbers:
- `range(5)` → 0, 1, 2, 3, 4
- `range(1, 6)` → 1, 2, 3, 4, 5
- `range(0, 10, 2)` → 0, 2, 4, 6, 8

````python
# for loop over a list
cities = ["Mumbai", "Delhi", "Bangalore", "Pune"]

for city in cities:
    # 'city' is a LOOP VARIABLE — it takes the value of each item in turn
    # First iteration: city = "Mumbai"
    # Second iteration: city = "Delhi"
    # ...and so on
    print(f"City: {city}")
````

````python
# range() — loop a specific number of times
for i in range(5):
    print(i, end=" ")   # end=" " means don't start a new line, print a space instead
print()  # print a blank line after

for i in range(1, 11):  # 1 to 10 (stop=11 is not included)
    print(i, end=" ")
print()

for i in range(0, 20, 3):  # 0, 3, 6, 9, 12, 15, 18 (step of 3)
    print(i, end=" ")
print()
````

````python
# enumerate() — gives you index AND value at the same time
# Very useful when you need to know the position of each item
fruits = ["apple", "banana", "mango"]

# Without enumerate (old, awkward way):
for i in range(len(fruits)):
    print(i, fruits[i])

print()  # blank line

# With enumerate (clean, Pythonic way):
for index, fruit in enumerate(fruits):
    print(f"{index + 1}. {fruit}")  # index+1 so it starts from 1 not 0
````

````python
# while loop — runs as long as the condition is True
# Use when you don't know in advance how many times to loop
count = 0
while count < 5:
    print(f"Count is {count}")
    count += 1   # count = count + 1 (shorthand)
    # WARNING: if you forget this line, count never changes → infinite loop!

print("Done!")
````

````python
# break and continue — control flow inside loops
# break    → exit the loop immediately
# continue → skip the rest of this iteration, go to the next one

for num in range(10):
    if num == 3:
        continue   # skip 3, jump to num=4
    if num == 7:
        break      # stop the entire loop at 7
    print(num, end=" ")
print()  # Output: 0 1 2 4 5 6
````

## Section 4 — Conditionals (if / elif / else)

**Conditionals** let your code make decisions.

- `if` — runs if the condition is True
- `elif` — (else if) runs if the previous condition was False and this one is True
- `else` — runs if ALL conditions above were False

Python uses **indentation** (spaces/tabs) to show which code belongs inside the block.
This is NOT optional — wrong indentation causes errors.

````python
# Change this value and re-run the cell to see different grades
score = 78

# Python checks each condition top to bottom and stops at the first True one
if score >= 90:
    grade = "A"    # only runs if score >= 90
elif score >= 80:
    grade = "B"    # only runs if score < 90 AND score >= 80
elif score >= 70:
    grade = "C"    # only runs if score < 80 AND score >= 70
elif score >= 60:
    grade = "D"
else:
    grade = "F"    # runs if ALL conditions above were False

print(f"Score {score} → Grade {grade}")
````

````python
# Comparison operators
x = 5
print(x == 5)   # True  — EQUALITY check (double == not single =)
print(x != 3)   # True  — NOT equal
print(x > 3)    # True  — greater than
print(x < 3)    # False — less than
print(x >= 5)   # True  — greater than or equal
print(x <= 4)   # False — less than or equal

# Logical operators: and, or, not
age = 20
has_id = True

# 'and' → BOTH conditions must be True
if age >= 18 and has_id:
    print("Entry allowed")

# 'or' → AT LEAST ONE condition must be True
city = "Delhi"
if city == "Mumbai" or city == "Delhi":
    print(f"{city} is a metro")

# 'in' → checks if a value exists in a list (very useful!)
allowed_cities = ["Mumbai", "Delhi", "Pune"]
if city in allowed_cities:
    print(f"{city} is allowed")
````

## Section 5 — Functions

A **function** is a reusable block of code with a name.
Instead of writing the same logic 10 times, you write it once as a function
and **call** it 10 times.

Structure:
```python
def function_name(parameter1, parameter2):
    # code
    return result
```

- `def` — keyword that starts a function definition
- **parameters** — inputs the function receives (local variable names)
- `return` — what the function sends back to the caller
- **calling** a function: `result = function_name(value1, value2)`

````python
# Define a function — this just DESCRIBES what it does, doesn't run yet
def greet(name):
    # 'name' is a parameter — it receives whatever value the caller passes in
    message = f"Hello, {name}!"
    return message   # sends the result back to whoever called this function

# Call the function — NOW it actually runs
result1 = greet("Aadit")   # 'Aadit' is the ARGUMENT (actual value passed in)
result2 = greet("Priya")

print(result1)
print(result2)
````

````python
# Default parameter values
# If the caller doesn't provide passing_score, it uses 60 automatically
def calculate_grade(score, passing_score=60):
    if score >= passing_score:
        return "Pass"
    else:
        return "Fail"

print(calculate_grade(75))        # uses default → passing_score=60 → Pass
print(calculate_grade(45))        # 45 < 60 → Fail
print(calculate_grade(45, 40))    # custom passing score → 45 >= 40 → Pass
````

````python
# Returning multiple values — Python packs them into a TUPLE
# A tuple is like a list but immutable (can't be changed after creation)
# Syntax: (value1, value2, value3)
def get_stats(numbers):
    total = sum(numbers)
    average = total / len(numbers)
    return total, average, min(numbers), max(numbers)
    # Python packs these 4 values into a tuple: (803, 80.3, 55, 98)

scores = [88, 92, 75, 95, 60]

# UNPACKING — assign each value of the tuple to a separate variable
total, avg, lo, hi = get_stats(scores)

print(f"Total: {total}")
print(f"Average: {avg:.1f}")  # :.1f = format as float with 1 decimal place
print(f"Range: {lo} to {hi}")
````

## Section 6 — Dictionaries

A **dictionary** stores **key-value pairs**, inside curly braces `{ }`.
Think of it like a real dictionary: you look up a word (key) to get its definition (value).

```python
student = {"name": "Aadit", "score": 88}
#            key     value     key    value
```

- Keys are usually strings, values can be anything
- Look up a value: `student["name"]` or `student.get("name")`
- `.get()` is safer — returns `None` (or a default) if the key doesn't exist,
  instead of crashing

**Lists of dicts** — the most common data structure in data science.  
Each dict = one row. The keys = column names. This is exactly how CSV data loads.

````python
# A dictionary — represents one student
student = {
    "name": "Aadit",
    "age": 18,
    "score": 88,
    "city": "Mumbai"
}

# Access by key
print(student["name"])             # Aadit — crashes if key doesn't exist
print(student.get("score"))        # 88 — safe version
print(student.get("email", "N/A")) # N/A — safe with default value

# Modify
student["score"] = 92              # update existing key
student["grade"] = "A"             # add NEW key (dicts grow dynamically)

# Remove
del student["grade"]               # delete a key

print(student)
````

````python
# Looping over a dictionary

# Loop over keys only
for key in student:
    print(key)

print()

# Loop over both key AND value (most useful)
for key, value in student.items():
    print(f"  {key}: {value}")

# Just get lists of keys or values
print("Keys:", list(student.keys()))
print("Values:", list(student.values()))
````

````python
# LIST OF DICTS — this is how CSV data looks after you load it with read_csv()
# Each dict is one row. The keys are column names.
students = [
    {"name": "Aadit",  "score": 88, "city": "Mumbai"},
    {"name": "Priya",  "score": 92, "city": "Delhi"},
    {"name": "Rohan",  "score": 75, "city": "Bangalore"},
    {"name": "Sneha",  "score": 95, "city": "Pune"},
]

# Loop and print all students
for s in students:
    print(f"{s['name']}: {s['score']}")

# Find highest scorer using max() with a 'key' function
# key=lambda s: s["score"]  →  'for each student, use their score as the sort key'
# lambda is a tiny anonymous one-line function: lambda x: x["score"] = def f(x): return x["score"]
best = max(students, key=lambda s: s["score"])
print(f"\nTop scorer: {best['name']} with {best['score']}")

# Filter: list comprehension to get only Mumbai students
mumbai = [s for s in students if s["city"] == "Mumbai"]
print(f"Mumbai students: {mumbai}")
````

## Section 7 — Working with JSON

**JSON** (JavaScript Object Notation) is a text format for storing structured data.
It's used everywhere — APIs, config files, databases.

JSON looks almost identical to Python dicts and lists — the key difference is:
- JSON is just **text** (a string or a file)
- A Python dict is a **live object in memory**

Python's `json` module converts between them:
- `json.dumps(obj)` → Python object **to** JSON string (`s` = string)
- `json.loads(string)` → JSON string **to** Python object (`s` = string)
- `json.dump(obj, file)` → Python object **to** JSON file
- `json.load(file)` → JSON file **to** Python object

````python
import json   # must import before using — json is a built-in module

# Python dict → JSON string
student = {
    "name": "Aadit",
    "age": 18,
    "subjects": ["Maths", "Science", "Python"],  # list inside dict is fine
    "scores": {"maths": 90, "science": 85}        # dict inside dict too
}

json_string = json.dumps(student, indent=2)  # indent=2 makes it pretty-printed
print(type(json_string))  # <class 'str'> — it's just text now
print(json_string)
````

````python
# JSON string → Python dict
# Notice: JSON uses true/false (lowercase), Python uses True/False (capital)
# json.loads() converts them automatically
json_data = '{"name": "Priya", "score": 92, "passed": true}'
parsed = json.loads(json_data)

print(type(parsed))           # <class 'dict'> — now it's a real Python dict
print(parsed["name"])         # Priya
print(type(parsed["passed"])) # <class 'bool'> — JSON true → Python True
````

````python
# Save to a JSON file and read it back
# 'with open(...)' = context manager — automatically closes the file when done
data_to_save = {
    "week": 1,
    "topic": "Python Fundamentals",
    "students": [
        {"name": "Aadit", "score": 88},
        {"name": "Priya", "score": 92},
    ]
}

# Write to file — "w" mode means write (creates file or overwrites)
with open("../data/sample_output.json", "w") as f:
    json.dump(data_to_save, f, indent=2)
print("Saved!")

# Read from file — "r" mode is default, no need to specify
with open("../data/sample_output.json") as f:
    loaded = json.load(f)

print("Loaded topic:", loaded["topic"])
print("First student:", loaded["students"][0])  # [0] = first item of the list
````

## Section 8 — Reading & Writing CSV Files

**CSV files** are just plain text files where:
- The first line is the **header** (column names)
- Each following line is **one row of data**
- Values are separated by **commas**

Open `data/students.csv` in Notepad to see this directly — it's just text.

Python's `csv` module reads this text and converts it into Python structures:
- `csv.DictReader` → reads each row as a **dict** (column name → value)
- `csv.DictWriter` → writes a list of dicts back to a CSV file

**Important:** Every value from `DictReader` comes back as a **string** (text).  
Even numbers like `88` come back as `"88"`. You must convert: `int(row["score"])`

````python
import csv  # built-in module

# Read students.csv
# 'with open(...)' — always use this for files. It auto-closes when the block ends.
# newline="" is required by csv module (it handles line endings itself)
students = []
with open("../data/students.csv", newline="") as f:
    reader = csv.DictReader(f)   # reads first row as column headers
    for row in reader:
        students.append(dict(row))   # convert from OrderedDict to plain dict

print(f"Loaded {len(students)} students")
print("First row:", students[0])
print("Type of score:", type(students[0]["score"]))  # str! not int
````

````python
# Process data — convert score to int, add a letter grade
def assign_grade(score):
    score = int(score)   # convert string "88" to integer 88
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

for student in students:
    student["score"] = int(student["score"])            # convert in-place
    student["grade_letter"] = assign_grade(student["score"])  # add new column

for s in students:
    print(f"  {s['name']:<10} Score: {s['score']}  Grade: {s['grade_letter']}")
    # :<10 = left-align in 10-character wide space
````

````python
# Write enriched data to a NEW CSV file
# 'w' mode creates the file (or overwrites it if it already exists)
with open("../data/students_with_grades.csv", "w", newline="") as f:
    fieldnames = ["name", "age", "grade", "score", "city", "grade_letter"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()    # writes the header line: name,age,grade,...
    writer.writerows(students)  # writes all the data rows

print("Saved students_with_grades.csv")
print("Check the data/ folder — the file is there!")
````
